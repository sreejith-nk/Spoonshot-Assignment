### Method based on presence of exact match

```python
hashtab={}
for word in word_lst:
  lst=word.split()
  j=0
  for char in lst:
    if(char.lower() in article.lower()):
      if word in hashtab:
        hashtab[word]+=1/len(lst)  ## If all were present this will add upto one
      else:
        hashtab[word]=1/len(lst)
```

```python
ranking=dict(sorted(hashtab.items(),reverse=True,key=lambda item: item[1])) ##Sorting the dictionary to obtain a ranking
```

###Matching using similarity measures

* Using the close matches module from difflib (I have tried some other measures as well but this seems to give the best results)

```python
article_lst=article.split()
```

```python
from difflib import get_close_matches
hashtab_2={}
for word_art in article_lst:
  for word in word_lst:
    if len(get_close_matches(word_art,word.split(),n=1,cutoff=0.7))> 0:##The cuttoff parameter decides what kind of similarity we can allow
      if word in hashtab_2:
        hashtab_2[word]+=1/len(word.split())  ## If all were present this will add upto one
      else:
        hashtab_2[word]=1/len(word.split())
```

```python
ranking2=dict(sorted(hashtab_2.items(),reverse=True,key=lambda item: item[1])) ##Sorting the dictionary to obtain a ranking
```
